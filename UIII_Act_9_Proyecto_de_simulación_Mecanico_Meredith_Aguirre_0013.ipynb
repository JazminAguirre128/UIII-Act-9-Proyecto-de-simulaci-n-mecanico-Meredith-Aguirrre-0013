```python
# --- ASISTENTE IA PARA MECÁNICOS ---
print("==========================================")
print("   BIENVENIDO AL ASISTENTE IA DE TALLER   ")
print("==========================================")

# La lista comienza vacía (almacenará las herramientas del mecánico)
herramientas = []

opcion = 0

# Estructura repetitiva para el menú interactivo
while opcion != 4:
    print("\n--- MENÚ PRINCIPAL ---")
    print("1. Agregar herramienta al carro de taller")
    print("2. Diagnóstico IA de falla mecánica")
    print("3. Ver estado del carro de herramientas")
    print("4. Salir del sistema")

    try:
        opcion = int(input("\nSeleccione una opción (1-4): "))
    except ValueError:
        print("Por favor, ingrese un número válido.")
        continue

    # Opción 1: Gestión de la Lista (Memoria de la IA)
    if opcion == 1:
        nueva_herramienta = input("Ingrese el nombre de la herramienta: ")
        # Se agrega un elemento a la lista utilizando append
        herramientas.append(nueva_herramienta)
        print("IA: Herramienta guardada con éxito.")

    # Opción 2: Lógica de Inteligencia Artificial (Sistema Experto)
    elif opcion == 2:
        print("\n--- ESCANEO DE FALLA MECÁNICA ---")

        # Primero, la IA verifica si hay herramientas en la lista para trabajar
        if herramientas == []:
            print("IA ALERTA: Carro vacío. No puedes reparar sin herramientas.")
        else:
            try:
                temperatura = float(input("Ingrese la temperatura del motor (°C): "))
                ruido_metalico = input("¿Se escucha un golpeteo metálico en el motor? (si/no): ").lower().strip()
                fuga_liquido = input("¿Hay presencia de fugas visibles (aceite/anticongelante)? (si/no): ").lower().strip()

                # Lógica de decisión de la IA basada en operadores lógicos y relacionales
                if temperatura > 105 and fuga_liquido == "si":
                    print("\nIA DECISIÓN: PELIGRO DE SOBRECALENTAMIENTO CRÍTICO.")
                    print("Acción: Apagar el motor de inmediato. Riesgo de soplar la junta de la culata.")
                elif temperatura > 105 and ruido_metalico == "si":
                    print("\nIA DECISIÓN: RIESGO DE MOTOR DESBIELADO / DESGASTE SEVERO.")
                    print("Sugerencia: Revisar presión de aceite y nivel de lubricante antes de volver a encender.")
                elif temperatura < 40 and ruido_metalico == "si":
                    print("\nIA DECISIÓN: FALLA EN FRÍO / POSIBLE PROBLEMA DE PUNTERÍAS O TAQUÉS.")
                    print("Sugerencia: Inspeccionar el sistema de distribución con tus herramientas.")
                else:
                    print("\nIA DECISIÓN: PARÁMETROS OPERATIVOS NORMALES.")
                    print("Sugerencia: Puedes proceder con el mantenimiento de rutina (cambio de filtros y fluidos).")
            except ValueError:
                print("Error: Ingrese un valor numérico válido para la temperatura.")

    # Opción 3: Visualización de la Lista
    elif opcion == 3:
        print("\n--- ESTADO DEL CARRO DE HERRAMIENTAS ---")
        if herramientas == []:
            print("El carro de herramientas está vacío actualmente.")
        else:
            # Mostramos el contenido de la lista directamente
            print("Herramientas disponibles:", herramientas)
            print("Total de items:", len(herramientas))

    # Opción 4: Salida
    elif opcion == 4:
        print("\nCerrando sesión de IA. ¡Trabaja con seguridad en el taller, mecánico!")

    # Manejo de entradas no válidas
    else:
        print("Opción no válida. Intenta de nuevo.")

print("------------------------------------------")
print("Simulación de Proceso Lógico finalizada.")
print("Programa realizado por Mereedith Aguirre NL = 2")
```
